```python
#common imports 
import numpy as np
import pandas as pd
import math
import random
import matplotlib.pyplot as plt

#Text processing libraries for classification, tokenization, stemming, tagging, parsing 
#and semantic reasoning
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer 
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer

#Sequence of characters used to check whether a pattern exists in a given text (string) or not
import re

#Allows to use the operating system dependent functionalities and to interact with the underlying operating
#system.
import os
LESSON_ROOT_DIR = "../"


DATA_PATH = os.path.join(LESSON_ROOT_DIR, "data")
os.makedirs(DATA_PATH, exist_ok=True)

RESULTS_PATH = os.path.join(LESSON_ROOT_DIR, "results")
os.makedirs(RESULTS_PATH, exist_ok=True)

IMAGES_PATH = os.path.join(LESSON_ROOT_DIR, "images")
os.makedirs(IMAGES_PATH, exist_ok=True)

#XGboost Model 
import xgboost as xgb

#Gensim is a Python library for topic modelling, document indexing and similarity retrieval 
#with large corpora
import gensim
import gensim.models
import gensim.downloader
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
from gensim.test.utils import datapath
from gensim import utils

#Scikit-learn : tools for predictive data analysis
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn import datasets, linear_model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

#To process textual data.
from textblob import TextBlob as tb


import matplotlib.pyplot as plt

#Define square distance
def sqdist(vector):
    return sum(x*x for x in vector)

def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # return string  
    return (str1.join(s))


def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob.words)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)


```
